```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
import os

plt.style.use('seaborn')
np.__version__
```

```python
data = np.load('./Data/colors.pkl', allow_pickle=True)

data.shape
```

```python
plt.scatter(x = data[:, 0], y = data[:, 1])
plt.title('Scatter plot')
```

```python
from sklearn.mixture import GaussianMixture

gm = GaussianMixture(n_components=7).fit(data)

gm.predict(data)  
    
```

```python
def generate_data(n_components, means, covariances):
    generated_data = []
    # for each cluster generate 50 data points with corresponding mean and covariance using 
    # numpy multivariate normal pdf function.
    
    for i in range(n_components):
        generated_data.extend(np.random.multivariate_normal(means[i], covariances[i], 50).tolist())
    
    return np.array(generated_data)    
        
```

```python
generated_data = generate_data(7, gm.means_, gm.covariances_)
```

```python
plt.scatter(generated_data[:, 0], generated_data[:, 1])
plt.title('Generated clusters')
```
